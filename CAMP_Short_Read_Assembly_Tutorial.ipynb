#**Pipeline Setup**

##**Remove the default `sample_data` directory.**

```python
%%bash
rm -r sample_data/
```

##**Install the package manager Mamba (faster) via the package manager Miniconda (slower).**

Install Miniconda.

```python
%%bash
wget https://repo.anaconda.com/miniconda/Miniconda3-py39_4.10.3-Linux-x86_64.sh &>/dev/null
chmod u+x Miniconda3-py39_4.10.3-Linux-x86_64.sh
./Miniconda3-py39_4.10.3-Linux-x86_64.sh -b -f -p /usr/local &>/dev/null
rm Miniconda3-py39_4.10.3-Linux-x86_64.sh
```

Install Mamba into the base environment.

```python
%%bash
conda install -c conda-forge mamba --yes &>/dev/null
```

#**Short Read Assembly**

##**Download the Short Read Assembly module and the included test data.**

```python
%%bash
git clone https://github.com/MetaSUB-CAMP/camp_short-read-assembly.git
```

##**Install all the programs of the CAMP `short-read-assembly` module into a new Conda environment named `campass` as listed in `short-read-assembly.yaml` .**

```python
%%bash
mamba env create -f camp_short-read-assembly/configs/conda/short-read-assembly.yaml
```

##**Install the workflow management system Snakemake in the `short-read-assembly` environment to manage the pipeline.**

Installing Snakemake into the same directory resolves a dependecy error.

```python
%%bash 
source activate campass
mamba install -c bioconda -c conda-forge snakemake=7.18.1 --yes
```

#**Project Setup**

Create the working directory, `shrtRdAs` (short-Read-Assembly) where output files and logs will be stored.

```python
%%bash
mkdir shrtRdAs/
```

Remake the `samples.csv` to change the file path, remove a reference to `camp_short-read-quality-control`, and change reverse reads to `zymo_pos_cntrl.r2.fq.gz`.

```python
%%bash
cat > samples.csv
,illumina_fwd,illumina_rev
zymotestsample,/content/camp_short-read-assembly/test_data/zymo_pos_cntrl.r1.fq.gz,/content/camp_short-read-assembly/test_data/zymo_pos_cntrl.r2.fq.gz
```

#**Run the Short Read Assembly module on the included example data.**

Run the Short Read Assembly module on `zymo_pos_cntrl.r1.fq.gz` and `zymo_pos_cntrl.r2.fq.gz`.

```python
%%bash
source activate campass
python3 camp_short-read-assembly/workflow/short-read-assembly.py -w /content/camp_short-read-assembly/workflow/Snakefile -d /content/shrtRdAs/ -s /content/samples.csv
```
